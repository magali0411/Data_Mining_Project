# Etape 1 : Rechercher les meilleurs pré-traitements et les meilleurs classifieurs

## *Pré-traitement des datas*

#### Tokenisation

```python
from nltk import sent_tokenize
from nltk.tokenize import word_tokenize
```

Découpage des commentaires en phrases

```python
for commentaires in liste :
    sent_tokenize(commentaires)

display(liste[1])
```

Découpage des commentaires en mot

```python
Lmot = []

for commentaires in liste :
    Lmot.append(word_tokenize(commentaires))

display(Lmot[0])

#Problème : ponctuation
```

### Nettoyage des données

Mise en minuscule pour faciliter la detection des mots

```python
L_mot_lower={}
i = 1 
for com in Lmot :
    for w in com :
        if i in L_mot_lower.keys() :
            L_mot_lower[i]+= w.lower()
        else :
            L_mot_lower[i] = w.lower()
    i +=1
```

```python
Lil = {}
i=1
for w in Lmot :
    Lil[i]=[]
    for ww in w :
        Lil[i].append(ww.lower())
    i +=1
        
        
display(Lil)
```

Suppression de tout ce qui n'est pas un caractère alphabétique ou numérique

```python
test = []

for commentaire in Lmot :
    for word in commentaire :
        if word.isalpha() :
            test.append(word)
print(test)
```

Suppresion des stop words

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

### *Extraction de features*

'Named Entity Rocognition' permet de recuperer des noms propres dans un texte. <br/>
'Extraction de Relation' pour modeliser les relations entre les mots. <br/>

### *TF/IDF : Fréquence d'apparition*

IDF : Inverse Document Frequency  <br/>
Si un mot est présent dans tous les commentaires : Peu d'importance *Idf = 0* <br/>
Si il apparait souvent, il peut caractériser un commentaire.  <br/>
<br/>
On peut se baser sur des corpus, ensemble de mots déjà lammtizé et stemmé qui correspondront au vocabulaire qu'on attend! Ici critique, cinéma, film (__A chercher__) ==> Matrice qui fait du sens

## Graph of Words

Avec la méthode des bag of words on suppose que tous les term son equi-pondérés et qu'ils sont tous indépendants. <br/>
La méthode par graphe apporte justement des relations entre mots, de l'ordre et une notion de distances personnalisées.<br/>
Notion de poid : Nombre d'arrete venant sur le mot. <br/>
*Un nouveau context = 1 nouvelle arrete*
__=> Nouvelle mesure du TF!! = Meilleure métrique TF IDF__ <br/>
